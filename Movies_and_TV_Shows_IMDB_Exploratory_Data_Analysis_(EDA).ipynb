## Data Preparation and Cleaning

Loaded the dataset into a data frame using Pandas. Then, I will explore basic information about the data such as the number of rows & columns, ranges of values etc. Next, I will handle the missing data by replacing or removing NaN values.



```python
raw_df=pd.read_csv('imdb-movie-rating/imdb.csv')
raw_df
```

```python
raw_df.shape
```

```python
raw_df.info()
```

```python
raw_df.describe()
```

```python
raw_df.Certificate = raw_df.Certificate.fillna('Not Rated')
raw_df
```

Below dropping the column Metascore. It not relevant to this study 

```python
raw_copy_df=raw_df.copy()
raw_copy_df.drop(columns=['Metascore'], inplace=True)
raw_copy_df.info()
```

I will be dropping all the nan values.

```python
imdb_movies_df=raw_copy_df.dropna()
imdb_movies_df
```

Below, I will be changing the name of a column.

```python
imdb_movies_df.rename(columns = {'Gross_in_$_M':'Gross_M'},inplace=True)
imdb_movies_df
```

```python
imdb_movies_df.info()
```

Let's understand the data.

```python
imdb_movies_df.describe()
```

```python
imdb_movies_df.head(10)
```

```python
imdb_movies_df.tail(10)
```

```python
imdb_movies_df.sample(10)
```

Below, I am grouping by the dataset.

```python
imdb_movies_year_df = imdb_movies_df.groupby('Year')['Gross_M'].sum()
imdb_movies_year_df
```

```python
imdb_movies_cert_df = imdb_movies_df.groupby('Certificate')[['Runtime_in_min','Rating_from_10']].mean()
imdb_movies_cert_df
```

## Exploratory Analysis and Visualization

Here, I compute the mean, sum, range and other interesting statistics for numeric columns. We explore the distributions of numeric columns using histograms etc. Further, we have analyzed the relationship between columns using scatter plots, pie charts etc.


Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
plt.title('IMDB MOVIES RUNTIME')
plt.hist(imdb_movies_df.Runtime_in_min, bins=np.arange(40,250, 10));
plt.ylabel('Runtime')
```

```python
plt.title('IMDB MOVIES RATING')
plt.hist(imdb_movies_cert_df.Rating_from_10, bins=np.arange(7.6, 8.6, 0.05), color='RED');
plt.xlabel('Rating')
```

Below, I have plotted the distribution of movies according to ratings out of 10. We notice that no movies have rating less than 7, so, to see the histogram clearly we update the range of histogram and plot in range 7 to 10.

```python
plt.title("Distribution of movies (rating based)")
plt.hist(imdb_movies_df.Rating_from_10, bins=np.arange(1, 10, 0.5), color='Black');
```

```python
plt.title("Distribution of movies (rating based)")
plt.hist(imdb_movies_df.Rating_from_10,bins=np.arange(7, 10, 0.2), color='Green');
```

In the following plots, I have plotted the duration distribution of movies and rating distribution of movies.

```python
sns.displot(imdb_movies_df.Rating_from_10, color='blue').set(title='Distribution of Movies Ratings')
```

```python
sns.displot(imdb_movies_df.Runtime_in_min, color='black').set(title='Duration Distribution of Movies')
```

Considering the following two plots corresponding to data for imdb_movies_cert_df and scatter plot is better at depicting the relation.

```python
plt.figure(figsize=(14,8))
plt.plot(imdb_movies_cert_df.Rating_from_10,imdb_movies_cert_df.Runtime_in_min)
plt.xlabel('Runtime (in min)')
plt.ylabel('Gross (in $M)');
```

```python
plt.figure(figsize=(14,8))
sns.scatterplot(x=imdb_movies_cert_df.Rating_from_10, y=imdb_movies_cert_df.Runtime_in_min);
```

Now, I plot  the simple plot and scatter plot for the data including Rating and Runtime and observe that the scatter plot is easy to analyse.

```python
plt.figure(figsize=(14,8))
plt.plot(imdb_movies_df.Rating_from_10,imdb_movies_df.Runtime_in_min,color='Orange')
plt.xlabel('Runtime (in min)')
plt.ylabel('Gross (in $M)');
```

```python
plt.figure(figsize=(14,8))
sns.scatterplot(x=imdb_movies_df.Rating_from_10, y=imdb_movies_df.Runtime_in_min, color='Brown');
```
